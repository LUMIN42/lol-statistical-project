I am comparing winrate on teams with a skewed adaptive damage split (4+ physical or 4+ magical) versus teams with a balanced split.

The question here is whether having a skewed damage composition is associated with lower win rates.

I am going to perform a permutation test with a 5 / 3 % significance threshold.
The 3 is a part of Bonferroni's correction since I am testing 3 separate hypotheses.

## Load the aggregated data

```python
import csv

rows = []

with open("ap_ad_split.csv", "r", encoding="utf-8") as f:
  reader = csv.reader(f)

  next(reader)  # skip header

  for win, skewed in reader:
    rows.append((win == "True", skewed == "True"))
```

## Compare win rates

```python
mixed_sample = [win for win, _ in rows]
mixed_winrate = sum(mixed_sample) / len(mixed_sample)
print(f"{mixed_winrate=:.2f} (sanity check, should be exactly 50 %)")

print()

skewed_sample = [win for win, skewed in rows if skewed]
skewed_winrate = sum(skewed_sample) / len(skewed_sample)
print(f"{skewed_winrate=:.2f}")
print(f"{len(skewed_sample)=}")

print()

unskewed_sample = [win for win, skewed in rows if not skewed]
unskewed_winrate = sum(unskewed_sample) / len(unskewed_sample)
print(f"{unskewed_winrate=:.2f}")
print(f"{len(unskewed_sample)=}")
```

Given that skewed winrate is higher than unskewed, we have definitely not disproven the null hypothesis.
